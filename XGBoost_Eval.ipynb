```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import os

%matplotlib inline
```

```python
# Loading  data
df = pd.read_csv('UCI_Credit_Card_balanced.csv')
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df.drop('Y',axis=1))
```

```python
scaled_features = scaler.fit_transform(df.drop('Y',axis=1))
```

Splitting labels and features

```python
X = pd.DataFrame(scaled_features,columns=df.columns[:-1]).values
y = df['Y'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=69)
```

```python
from xgboost import XGBClassifier
```

```python
min_child_weight = 1
gamma = 0.5
subsample = 0.6
colsample_bytree = 1.0
max_depth = 5
```

```python
xgb = XGBClassifier(learning_rate=0.02, n_estimators=600, objective='binary:logistic', nthread=1,
                    min_child_weight=min_child_weight, gamma=gamma, subsample=subsample, colsample_bytree=colsample_bytree, max_depth=max_depth)                
```

```python
from sklearn.metrics import plot_confusion_matrix, cohen_kappa_score, plot_roc_curve, accuracy_score, roc_auc_score, precision_recall_curve, confusion_matrix, roc_curve, auc
import shap
```

```python
t0 = datetime.now()
xgb.fit(X_train,y_train)
t1 = datetime.now()
print("Training Time: ", (t1 - t0))
```

```python
predictions = xgb.predict(X_test)
```

```python
plt.figure(figsize=(7, 5))
ax = plot_confusion_matrix(xgb,X_test,y_test,cmap=plt.cm.Blues)

plt.tight_layout()
plt.savefig(r'C:\Users\alanw\OneDrive\Varsity\Research\Paper Itself\LaTeX\images\xgb_conf.png')
plt.show()
```

```python
plt.figure(figsize=(7, 5))
ax = plot_roc_curve(xgb,X_test,y_test)
x = np.linspace(0, 1, 100)
ax = plt.plot(x, x, linestyle='--',color='black')

plt.tight_layout()
plt.savefig(r'C:\Users\alanw\OneDrive\Varsity\Research\Paper Itself\LaTeX\images\xgb_roc.png')
plt.show()
```

```python
kappa = cohen_kappa_score(y_test, predictions)
print("Cohen's Kappa: %.2f" % (kappa))
```

```python
shap_values = shap.TreeExplainer(xgb).shap_values(X_train)

plt.figure(figsize=(7, 5))
ax = shap.summary_plot(shap_values, X_train, plot_type="bar", show=False)

plt.tight_layout()
plt.savefig(r'C:\Users\alanw\OneDrive\Varsity\Research\Paper Itself\LaTeX\images\xgb_shap.png')
plt.show()
```
